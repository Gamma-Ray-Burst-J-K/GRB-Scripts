```python
def power_law1(x, N, a): 
    '''simple power law 
    L = Nx^-a'''
    return N*np.power(x, a)
```

```python
def power_law3(x,N,a):
    '''simple power law in log space
    log(L) = log(N) - alogx'''
    return (a)*np.log10(x) + np.log10(N) 
```

```python
'''BOOTSTRAP MONTE CARLO'''
long_grbs = []
boot_trials=10000

for g in ras_data:
    if g[2] >= 2:
        long_grbs.append(g)
        
        
# ts_rank, ts_null, U_rank, U_null, per_rank, per_null, peak_rank, peak_null = [], [], [], [], [], [], [], []
rank, null_hyp, Lx_error_per = [], [], []

for i in range(boot_trials):
    Lx_data, ax_data, Lx_error, ax_error = [], [], [], []

    #selecting 18 (final sample size for GeV) random integer values to run bootstrap on
    selected = random.randint(0,len(long_grbs),len(grb_name))
    
    for j in selected:
        Lx_data.append(long_grbs[j][5])
        Lx_error.append(long_grbs[j][6])
        ax_data.append(long_grbs[j][3])
        ax_error.append(long_grbs[j][4])
        #Lx_error_per.append(long_grbs[j][6]/long_grbs[j][5] * 100)
        
    spearman = stats.spearmanr(np.log10(Lx_data), ax_data)
    rank.append(spearman[0])
    null_hyp.append(spearman[1])
    
    
    
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,10))

rank_median, null_median = np.median(rank), np.median(null_hyp)
rank_mean, null_mean = np.mean(rank), np.mean(null_hyp)
rank_std, null_std = np.std(rank), np.std(null_hyp)


#calculating 1,2 and 3 sigma errors
rank_quantile_159, rank_quantile_841 = np.quantile(rank, 0.159), np.quantile(rank, 0.841)
null_hyp_quantile_159, null_hyp_quantile_841 = np.quantile(null_hyp, 0.159), np.quantile(null_hyp, 0.841)
rank_1_sig = (abs(rank_quantile_841 - rank_mean) + abs(rank_quantile_159 - rank_mean))/2
null_1_sig = (abs(null_hyp_quantile_841 - null_mean) + abs(null_hyp_quantile_159 - null_mean))/2

rank_quantile_023, rank_quantile_977 = np.quantile(rank, 0.023), np.quantile(rank, 0.977)
null_hyp_quantile_023, null_hyp_quantile_977 = np.quantile(null_hyp, 0.023), np.quantile(null_hyp, 0.977)
rank_2_sig = (abs(rank_quantile_977 - rank_mean) + abs(rank_quantile_023 - rank_mean))/2
null_2_sig = (abs(null_hyp_quantile_977 - null_mean) + abs(null_hyp_quantile_023 - null_mean))/2

rank_quantile_002, rank_quantile_998 = np.quantile(rank, 0.002), np.quantile(rank, 0.998)
null_hyp_quantile_002, null_hyp_quantile_998 = np.quantile(null_hyp, 0.002), np.quantile(null_hyp, 0.998)
rank_3_sig = (abs(rank_quantile_998 - rank_mean) + abs(rank_quantile_002 - rank_mean))/2
null_3_sig = (abs(null_hyp_quantile_998 - null_mean) + abs(null_hyp_quantile_002 - null_mean))/2


#plotting distributions of Rsp and p
ax1.set_title(f'{boot_trials} trials - Spearman Coefficient Distribution', fontsize=15)
ax1.set_ylabel('Number of Trials', fontsize=15)
ax1.set_xlabel('Spearman Coefficient', fontsize=15)
ax1.hist(rank, fc=(0, 0, 1, 0.5), edgecolor='black')
ax1.axvline(rank_mean+rank_1_sig,color='red',label='1 sigma')
ax1.axvline(rank_mean-rank_1_sig,color='red')
ax1.axvline(rank_mean+rank_2_sig,color='green',label='2 sigma')
ax1.axvline(rank_mean-rank_2_sig,color='green')
ax1.axvline(rank_mean+rank_3_sig,color='black',label='3 sigma')
ax1.axvline(rank_mean-rank_3_sig,color='black')
ax1.legend()

ax2.set_title(f'{boot_trials} trials - Spearman Null Hypothesis Distribution', fontsize=15)
ax2.set_ylabel('Number of Trials', fontsize=15)
ax2.set_xlabel('Spearman Null Hypothesis', fontsize=15)
ax2.hist(null_hyp,  color='blue', fc=(0, 0, 1, 0.5), label='All', edgecolor='black')
# ax2.set_xscale('log')

print( rank_mean, rank_std)
print( null_mean, null_std)

```
